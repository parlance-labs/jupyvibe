## What the `patch` decorator does:

1. **Adds methods to existing classes**: The `@patch` decorator allows you to define a function and attach it as a method to an existing class, even if you don't have access to the original class definition.

2. **Syntax**: You use it like `@patch` above a function definition, and the first parameter of that function becomes the `self` parameter when it's attached to the class.

3. **Dynamic method addition**: This is particularly useful for:
   - Extending functionality of built-in classes
   - Adding methods to classes from external libraries
   - Organizing code by keeping related functionality together even when added to different classes

## Example usage pattern:

````python
@patch
def new_method(self:SomeExistingClass, param1, param2):
    # Function body here
    return something
````

This would add `new_method` to `SomeExistingClass`, so you could then call `instance.new_method(param1, param2)` on any instance of that class.

The patch decorator is a key feature of fastcore that enables a more flexible and modular approach to extending existing classes without modifying their original source code.

## Using Custom Functions as Tools

You can also define your own functions and expose them as tools.

````python
def count_lines(filepath: str) -> int:
    """Count the number of lines in a file."""
    with open(filepath) as f:
        return len(f.readlines())

def get_imports(filepath: str) -> list:
    """Extract all import statements from a Python file."""
    imports = []
    with open(filepath) as f:
        for line in f:
            line = line.strip()
            if line.startswith('import ') or line.startswith('from '):
                imports.append(line)
    return imports
````

## Working with Data Variables

The `$variable` syntax is especially useful for data analysis.

````python
# Sample sales data
sales_data = [
    {'product': 'Widget A', 'units': 150, 'price': 25.99},
    {'product': 'Widget B', 'units': 89, 'price': 45.50},
    {'product': 'Gadget X', 'units': 234, 'price': 12.99},
    {'product': 'Gadget Y', 'units': 67, 'price': 89.99},
]

def calculate_revenue(data: list) -> float:
    """Calculate total revenue from sales data."""
    return sum(item['units'] * item['price'] for item in data)

def top_seller(data: list) -> dict:
    """Find the product with the most units sold."""
    return max(data, key=lambda x: x['units'])
````

## Summary

ai-jup lets you:

1. **Reference variables** with `$variable_name` - the value is substituted into the prompt
2. **Expose functions as tools** with `&function_name` - the AI can call them during response
3. **See context** - the AI sees all preceding code cells and kernel state

This enables powerful workflows where you can:
- Explore codebases with file tools (`&view`, `&rg`)
- Analyze data with custom functions
- Let the AI use your domain-specific tools

### Syntax Comparison

| Solveit Syntax | ai-jup Syntax | Purpose |

| `` $`varname` `` | `$varname` | Reference variable value |
| `` &`funcname` `` | `&funcname` | Expose function as tool |
